```python
import gensim
from gensim.models import Word2Vec
from gensim.test.utils import common_texts, get_tmpfile
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format('../input/GoogleNews-vectors-negative300.bin', binary=True) 
```

```python
print(model['drink'])
```

```python
def avg_sentence_vector(words, model, num_features):
    featureVec = np.zeros((num_features,), dtype="float32")
    nwords = 0
    for word in words:
        featureVec = np.add(featureVec, model[word])

    return featureVec
```

```python
s1=['I','am','going','India']
f1=avg_sentence_vector(s1,model,300)
s2=['I','will','be','drinking','coffee']
f2=avg_sentence_vector(s2,model,300)
s4=['I','will','be','eating','coffee']
f4=avg_sentence_vector(s4,model,300)
s3=['I','am','going','Bharat']
f3=avg_sentence_vector(s3,model,300)
```

```python
print(f1)
```

```python
f1=f1.reshape((300,1))
f2=f2.reshape((300,1))
f3=f3.reshape((300,1))
f4=f4.reshape((300,1))
print(f1.shape)
```
